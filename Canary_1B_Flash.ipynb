```python
from pydub import AudioSegment
from IPython.display import Audio, display
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('Running inference on: ',device )
canary_model = EncDecMultiTaskModel.from_pretrained(model_name= 'nvidia/canary-1b-flash', map_location=device)
```

```python
!ls
```

```python
original_audio_path = "health-german.mp3"

# convert to compatible format: 16kHz, WAV, Mono
audio = AudioSegment.from_file(original_audio_path)
audio = audio.set_frame_rate(16000).set_channels(1)
convert_audio_path = "audio_16k.wav"
audio.export(convert_audio_path, format="wav")
print("Audio has been converted and saved at: ", convert_audio_path)
```

```python
display(Audio(convert_audio_path))
```

```python
transcript = canary_model.transcribe(
    audio=[convert_audio_path],
    source_lang = 'de',
    target_lang = 'de',
    pnc = 'True'
)
print('Transcription (German ASR with punctuation)')
print(transcript[0].text)
```

```python
transcript = canary_model.transcribe(
    audio=[convert_audio_path],
    source_lang = 'de',
    target_lang = 'en',
    pnc = 'True'
)
print('Transcription (German to Englsih Text)')
print(transcript[0].text)
```
